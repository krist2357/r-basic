```python
import numpy as np
import math
def raices(a,b,c):
    d=math.pow(b,2)-4*a*c
    if d < 0:
        z=abs(d)
        r1=(-b+math.sqrt(z)*1j)/(2*a)
        r2=(-b-math.sqrt(z)*1j)/(2*a)
    else:
        r1=(-b+math.sqrt(d))/(2*a)
        r2=(-b-math.sqrt(d))/(2*a)
    return r1, r2
```

```python
raices(3,1,-5)
```

```python
def primo(y):
    import numpy as np
    import math
    o=np.arange(y)
    
    print("Se crea el vector o:")
    print(o+1)
    q=np.arange(y)
    
    for x in o:
        r2=math.fmod(y,x+1)
        q[x]=r2
    
    print("Se crea un vector q con los residuos que resultan de la división del número ingresado y el vector o:")
    print(q)
    
    m=q[1:y-1]
    print("Se crea un vector m con los residuos del número ingresado y el vector o (Excepto para el 1 y si mismo):")
    print(m)
    n=len(m)
    s=np.arange(n)
    
    count=0
    for z in s:
        if m[z]==0:
            count=count+1 #Cuento cuantas veces hay un cero.
    print("")
    print("La cantidad de numeros divisibles para el número ingresado es: %s"%count)
    print("")
    if count!=0:
        print("El número ingresado no es un numero primo")
    else:
        print("El número ingresado es un numero primo")
```
